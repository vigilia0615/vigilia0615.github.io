4.分群k-means

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, datasets, metrics
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')

import pandas as pd
#讀入資料
enroll = pd.read_csv("dataset.csv")
enroll.head(450)
#設定自變量X，因變量y
x=np.array(enroll[['GRE Score','TOEFL Score','University Rating','SOP','LOR', 'CGPA','Research']])
y=np.array(enroll[['Chance of Admit ']])
# 迴圈
silhouette_avgs = []
ks = range(2, 11)
for k in ks:
    kmeans_fit = cluster.KMeans(n_clusters = k).fit(x)
    cluster_labels = kmeans_fit.labels_
    silhouette_avg = metrics.silhouette_score(x, cluster_labels)
    silhouette_avgs.append(silhouette_avg)

# 作圖並印出 k = 2 到 10 的績效
plt.bar(ks, silhouette_avgs)
plt.show()
print(silhouette_avgs)
```

```python
km = KMeans(n_clusters=2)  #K=2群
y_pred = km.fit_predict(x)
plt.figure(figsize=(10, 6))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x[:, 0], x[:, 1], c=y_pred) #C是第三維度 已顏色做維度
plt.show()

km.cluster_centers_ #各群中心點(X,Y)的位置



```
